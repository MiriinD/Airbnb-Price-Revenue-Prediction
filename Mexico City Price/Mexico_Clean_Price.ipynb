```python
def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.1)
    quartile3 = dataframe[variable].quantile(0.9)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit

def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = round(low_limit,0)
    dataframe.loc[(dataframe[variable] > up_limit), variable] = round(up_limit,0)
    
columns = ["latitude",
           "longitude",
           "accommodates",
           "bathrooms",'beds','price','availability_365', 'number_of_reviews']
for col in columns:
    replace_with_thresholds(numericals_df, col)
```

## Statistical Significance

It's important to test the statistical significance of the correlation coefficients to ensure that the observed relationship is not due to random chance. This can be done using tests such as the Pearson correlation significance test.

```python
from scipy.stats import pearsonr

# Assuming final_df is your DataFrame and it has been defined earlier

final_col = final_df.columns

for col in final_col:
    # Skip the 'Price' column to avoid calculating its correlation with itself
    if col == 'price':
        continue

    # Calculate the Pearson correlation coefficient and p-value
    corr_coefficient, p_value = pearsonr(final_df[col], final_df['price'])

    # Check if the p-value is less than or equal to 0.05
    if p_value > 0.05:
        # Print the column name, correlation coefficient, and p-value for significant correlations
        print(f'{col} Pearson Correlation Coefficient: {corr_coefficient:.3f}')
        print(f'{col} P-value: {p_value:.3e}')
```

- Correlation Coefficient (corr_coefficient): This value ranges from -1 to 1, where closer to 1 or -1 indicates a strong linear relationship, and closer to 0 indicates a weak or no linear relationship.
- P-value (p_value): This value helps you determine the significance of the correlation. A common threshold for significance is 0.05:
If the p-value is less than or equal to 0.05, the correlation is considered statistically significant, meaning there's strong evidence that the true correlation is not zero.
If the p-value is greater than 0.05, the correlation is not considered statistically significant, suggesting that the observed correlation might be due to random variation in the sample.

## Multicollinearity

Including variables that are highly correlated with each other (not just with the dependent variable) can lead to multicollinearity, which can skew the results of a linear regression model. It's often advisable to avoid including variables that have a very high correlation with one another.

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Assuming final_df is your DataFrame and it has been defined earlier, and contains only explanatory variables
X = add_constant(final_df)

# Calculate VIF for each explanatory variable
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Filter out features with VIF equal to or greater than 5
high_vif = vif_data[vif_data["VIF"] >= 5]

print(high_vif)
```

- A rule of thumb is that if VIF is 1, then the variable is not correlated with any other variables.
- A VIF between 1 and 5 suggests moderate correlation, but it's often not severe enough to require attention.
- A VIF greater than 5 or 10 indicates high multicollinearity that you may want to address by removing the variable or using techniques like Principal Component Analysis (PCA) or partial least squares regression.

## Practical Thresholds

As a rule of thumb, some analysts might consider variables with a correlation coefficient above a certain threshold, such as |0.2| or |0.3|, as potentially useful predictors. This threshold is arbitrary and should be adjusted based on the context of the analysis and the points mentioned above.

```python
import numpy as np  # Ensure numpy is imported

# Calculate correlation matrix
corr_matrix = final_df.corr().abs()

# Select upper triangle of correlation matrix to avoid double counting
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool_))

# Define a correlation coefficient threshold
corr_threshold = 0.7

# Find features with correlation greater than the threshold
high_corr_pairs = [(var1, var2) for var1 in upper_tri.columns for var2 in upper_tri.index if upper_tri.at[var2, var1] > corr_threshold]

print("Highly correlated pairs:", high_corr_pairs)
```

- Review and Understand: Understand the role of these variables in your analysis. Are they essential? Can the information they provide be captured by other variables?
- Remove or Combine Variables: Consider removing one of the variables in a highly correlated pair or combining variables to reduce multicollinearity.
- Regularization: Apply regularization methods (Lasso or Ridge regression) that can handle multicollinearity by penalizing the size of coefficients.
- Dimensionality Reduction: Use techniques like Principal Component Analysis (PCA) to create a set of uncorrelated predictors.
